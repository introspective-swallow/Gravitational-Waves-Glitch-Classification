```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Sampling

```python
t = np.linspace(0,5.5, 1000)
freq = 1
periode = 1 / freq
x = np.sin(2* math.pi*freq*t)
sr = 13*freq
T = 1 / sr
td = np.arange(0,5.5, T)
xd = np.sin(2* math.pi*freq*td)
fig, ax = plt.subplots()
ax.scatter(td, xd, color="black", s=30)
ax.plot(t, x, '--')
ax.set_yticks([])
ax.set_xticks([])
ax.axhline(y=0, color='black')
ax.grid(True, which='both')
```

# Aliasing

```python
t1 = 0
t2 = 8
t = np.linspace(t1,t2, 1000)

freq = 1/8
periode = 1 / freq
x = np.sin(2* math.pi*freq*t)

freq2 = -7 / 8
x2 = np.sin(2* math.pi*freq2*t)

sr = 1
T = 1 / sr

td = np.arange(t1,t2+1, T)
xd = np.sin(2* math.pi*freq*td)
fig, ax = plt.subplots()
ax.scatter(td, xd, color="black", s=30)
ax.plot(t, x, '--')
ax.plot(t, x2, '--', color="red")
ax.set_yticks([])
ax.set_xticks([])
ax.axhline(y=0, color='black')
ax.grid(True, which='both')
```

# DFT

```python
N = 25

n = np.arange(1000)
w0=5
x = np.cos(w0 * n)

xfft =np.fft.fft(x, 25)

plt.plot(xfft)
```

```python
L = 25
N=512
n = np.arange(L)
w0=math.pi / 2
x = np.cos(w0 * n)

xfft =np.fft.fft(x, N)

freq = 2*math.pi* np.fft.fftfreq(N)

plt.plot(np.fft.fftshift(freq), np.sqrt(xfft**2))
plt.xticks([-math.pi,-math.pi/2,0,math.pi/2,math.pi], ['-π', '-π/2', '0', 'π/2', 'π'])
```

```python
L = 25
N=512
n = np.arange(L)


# Hann window
w = 0.5 * (1 - np.cos(2*np.pi / (L-1) * n ))

x = np.cos(w0 * n)

x = x * w

xfft =np.fft.fft(x, N)

freq = 2*math.pi* np.fft.fftfreq(N)

plt.plot(np.fft.fftshift(freq), np.sqrt(xfft**2))
plt.xticks([-math.pi,-math.pi/2,0,math.pi/2,math.pi], ['-π', '-π/2', '0', 'π/2', 'π'])
```
